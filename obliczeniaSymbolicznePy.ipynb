```python
import sympy as sp

# Zmienne
h1, h2, F1, Fd = sp.symbols('h1 h2 F1 Fd', positive=True)

# Funkcje
f1 = (F1 + Fd - 23*sp.sqrt(h1)) / (0.7*h1)
f2 = (23*sp.sqrt(h1) - 30*sp.sqrt(h2)) / (1.35*h2**2)

# Punkt pracy
F1_val = 300
Fd_val = 200
h1_val = ( 1/23 * (F1_val + Fd_val) )**2
h2_val = ( 1/30 * (F1_val + Fd_val) )**2
point_of_operation = {h1: h1_val, h2: h2_val, F1: F1_val, Fd: Fd_val}

# Pochodne
df1_h1 = sp.diff(f1, h1)
df1_h2 = sp.diff(f1, h2)
df1_F1 = sp.diff(f1, F1)
df1_Fd = sp.diff(f1, Fd)

df2_h1 = sp.diff(f2, h1)
df2_h2 = sp.diff(f2, h2)
df2_F1 = sp.diff(f2, F1)
df2_Fd = sp.diff(f2, Fd)

print("Pochodne funkcji f1:")
print()
sp.pprint(df1_h1)
print()
sp.pprint(df1_h2)
print()
sp.pprint(df1_F1)
print()
sp.pprint(df1_Fd)
print("\nPochodne funkcji f2:")
print()
sp.pprint(df2_h1)
print()
sp.pprint(df2_h2)
print()
sp.pprint(df2_F1)
print()
sp.pprint(df2_Fd)
print()

# Macierze A i B liniaryzacji

A = sp.Matrix([[df1_h1.subs(point_of_operation), df1_h2.subs(point_of_operation)],
    [df2_h1.subs(point_of_operation), df2_h2.subs(point_of_operation)]])

B = sp.Matrix([[df1_F1.subs(point_of_operation), df1_Fd.subs(point_of_operation)],
    [df2_F1.subs(point_of_operation), df2_Fd.subs(point_of_operation)]])

print("Macierz A =")
print()
sp.pprint(A)
print()

print("Macierz B =")
print()
sp.pprint(B)
print()
```
